```python
import pandas as pd
import numpy as np

# =========================================================
# LOAD RATINGS
# =========================================================
ratings = pd.read_csv("../results/model_ratings_poisson_timeweighted.csv")

# =========================================================
# GLOBAL TEAM SCORE
# =========================================================
ratings["global_strength"] = ratings["attack"] - ratings["defense"]

# =========================================================
# RANKING
# =========================================================
ranking = ratings.sort_values("global_strength", ascending=False).reset_index(drop=True)

# =========================================================
# TOP 20
# =========================================================
print("\n===== TOP 20 TEAMS (GLOBAL STRENGTH) =====\n")
print(ranking.head(20)[["team", "attack", "defense", "global_strength"]])

# =========================================================
# BOTTOM 20
# =========================================================
print("\n===== BOTTOM 20 TEAMS =====\n")
print(ranking.tail(20)[["team", "attack", "defense", "global_strength"]])

# =========================================================
# OPTIONAL: VISUALIZATION
# =========================================================
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 8))
top = ranking.head(20)

plt.barh(top["team"][::-1], top["global_strength"][::-1])
plt.title("Top 20 teams - Poisson model strength")
plt.xlabel("Global strength (attack - defense)")
plt.show()
```
